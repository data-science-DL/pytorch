```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import matplotlib
from matplotlib import font_manager, rc
import platform
if platform.system()=="Windows":
    font_name=font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
matplotlib.rcParams['axes.unicode_minus']=False

import warnings
warnings.filterwarnings("ignore")
```

```python
score = [78.19, 56.47, 71.18, 69.83]
xlabel = ['Drift 전 모델', 'Drift 후 미대응', 'Drift 후 새 모델 생성', 'Drift 후 retraining']
```

```python


plt.figure(figsize=(12,12))
ax = sns.barplot(x=xlabel, y=score)
plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=18, labelcolor='black', top=True)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height+0.5, height, ha = 'center', size = 20)
plt.ylim([50,80])
plt.show()
```

```python
font1 = {
         'weight': 'bold',
         'size': 25
         }
x = [0,1,2,3,4]
score1 = [61.11, 69.83, 71.04, 72.06, 71.18]

plt.figure(figsize=(12,12))
plt.plot(score1, color='orange', marker='o', markersize=9, linewidth='7')

for i, v in enumerate(x):
    plt.text(v, score1[i]+0.3, str(score1[i]),
             fontsize=30,
             horizontalalignment='center',
             verticalalignment='bottom')

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=25, labelcolor='black', top=True)
plt.xticks([0,1,2,3,4], labels=['20%', '40%', '60%', '80%', '100%'])
plt.yticks([60, 65, 70, 75], labels=['60%', '65%', '70%', '75%'])
plt.xlabel('재학습률(%)', labelpad=15 ,fontdict=font1)
plt.ylabel('정확도(%)',  fontdict=font1)
plt.ylim([60,75])
plt.show()
```
